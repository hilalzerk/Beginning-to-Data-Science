## Variables 

A variable is used to store a value under a specific name.
The first step is to define the name of the variable. This name can be anything you choose, such as hilal, x, age, or numbers. However, it is important to ensure that the chosen name is not one of Python’s reserved keywords. For example, keywords like if, continue, and for are predefined by Python and therefore cannot be used as variable names.

```python
name = "Hilal"
age = "25"
learning = "Data Science"
```

## Data Types 

In Python, the most fundamental data types are as follows:

| Data Type | Description                    | Example          |
| --------- | ------------------------------ | ---------------- |
| **int**   | Integer numbers                | 10, 42           |
| **float** | Floating-point numbers         | 3.14, 2.0        |
| **str**   | Text / string values           | `"Hello"`        |
| **bool**  | Boolean (logical) values       | `True` / `False` |
| **list**  | Ordered collection of elements | `[1, 2, 3]`      |
| **tuple** | Immutable collection           | `(1, 2, 3)`      |
| **dict**  | Key–value data structure       | `{"a": 1}`       |
| **set**   | Collection of unique elements  | `{1, 2, 3}`      |
