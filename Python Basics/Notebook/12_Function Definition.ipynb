# 1. Function Definition (def)

Functions are code blocks that allow us to group repetitive operations under a single structure.

In data science workflows, functions are critically important for data cleaning, transformation, modeling, and automation.

```python
def greet(name):
    return f"Hello, {name}!"

print(greet("Hilal"))
```

```python
def students(name):
    notes = {"Ela": 40, "Ayşe": 60, "Ahmet": 85}
    return f"The grade of the student named {name}: {notes.get(name, 'Not found')}!"

print(students("Ela"))
```

```python
def student_status(name):
    grades = {"Ela": 70, "Ayşe": 45, "Ahmet": 90, "Mehmet": 55}
    score = grades.get(name)
    if score is None:
        return "Student not found"
    if score > 50:
        return f"{name} grade: {score}, Status: Passed"
    else:
        return f"{name} grade: {score}, Status: Failed"

print(student_status("Ela"))
```

```python
def max_min_grade():
    grades = {"Ela": 70, "Ayşe": 45, "Ahmet": 90, "Mehmet": 55}
    highest = max(grades, key=grades.get)
    lowest = min(grades, key=grades.get)
    return f"Highest: {highest} - {grades[highest]}\nLowest: {lowest} - {grades[lowest]}"

print(max_min_grade())
```
