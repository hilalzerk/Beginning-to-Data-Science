## 3. For Loops

For loops allow us to perform repetitive operations on data structures such as lists, strings, and dictionaries.

- Iterating over data
- Data cleaning operations
- Creating lists or dictionaries

```python
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit.upper())
```

```python
total = 0

for x in range(101):
    total += x
print(total)
```

```python
for x in range(5):
    print(x)
```

```python
total = 1
for i in range(5):
    total *=5
    
print(total)
```
