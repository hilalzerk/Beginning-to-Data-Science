## 4. While Loops

The loop continues as long as a condition is true.
If not handled carefully, it may result in an infinite loop.

```python
count = 0

while count < 3:
    print("Iteration:", count)
    count += 1
```

```python
s = "hey"
for c in s:
    print(c)
```

```python
n = len(s)
index = 0

while index < n:
        print(s[index])
        index += 1
```
