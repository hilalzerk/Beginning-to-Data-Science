## 5. Loop Control Statements

They are used to modify loop behavior:

- `break`: Terminates the loop completely
- `continue`: Skips the current iteration and continues the loop
- `pass`: Acts as a placeholder; performs no operation

```python
for num in [1, 2, 3, 4, 5]:
    if num == 3:
        continue  # pass the number 3
    if num == 5:
        break     # stop when it reaches number 5
    print(num)
```

```python
x = 0

while x < 10:
    
    x += 1
    
    if x == 8:
        continue
    
    print(x)
```
