## 2. Comparison and Logical Operators

They are frequently used in data filtering, data cleaning, and analysis steps.

Comparison operators:
== (equals)
!= (not equal)
<, >, >=, <=
Logical operators:
and
or
not

```python
3 == 4
```

```python
3 != 3
```

```python
4 < 20
```

```python
45 <= 34
```

```python
(45 <= 34) or (32 ==45 )
```

```python
(45 <= 34) & (32 ==45 )
```

```python
age = 25
income = 80000

print(age > 18 and income > 50000)   
print(age > 30 or income > 70000)   
print(not age > 30)                  
```
