## 2.2. Adding Elements to a List (append)

```python
numbers = [10, 20, 30] 
numbers.append(40)  # append() , adds a single element to the end of the list.
numbers
```

```python
numbers = [10, 20, 30] 
numbers.extend([50,60,70])  # extend() , adds multiple elements from another iterable.
numbers
```

```python
numbers = [10, 20, 30] 
numbers.insert(0,100)  # insert() , adds an element at a specific index.
numbers
```

```python
numbers = [10, 20, 30] 
numbers[0]= 100  # If the element at index 0 were "10", it would be removed and replaced with "100" when using direct assignment. However, when using insert(), no deletion occurs.
numbers
```

## 2.3. Removing Elements from a List (pop, remove)

```python
numbers = [10, 20, 30]
numbers.pop()     # This removes the last element from the list.
numbers
```

```python
numbers
```

```python
numbers.remove(20)
                  # Removes the first occurrence of the specified value from the list.
```

## 2.4. Slicing

```python
numbers = [10, 20, 30, 40, 50, 60]
```

```python
numbers[:2]  # 0 is the starting index (inclusive).
             # 2 is the ending index (exclusive).
```

```python
numbers[1:4] # 1 is the starting index (inclusive).
             # 4 is the ending index (exclusive).
```

## 2.6. Counting Elements in a List,Count()
In Python, the count() method is used to find how many times a specific value appears in a list.

```python
numbers = [10, 20, 30, 10, 40, 50, 10, 60, 30, 40]
numbers.count(10)
```

## 2.7. Sorting a List
In Python, lists can be sorted in ascending or descending order using built-in methods. Sorting is a fundamental operation in data manipulation, especially in data analysis and preprocessing.

Using sorted() – Returns a new sorted list without modifying the original list
Using .sort() – Sorts the list in place, modifying the original list

```python
numbers = [10, 20, 30, 10, 40, 50, 10, 60, 30, 40]
```

```python
sorted(numbers)
```
