A tuple in Python is similar to a list but immutable, meaning its elements cannot be changed, added, or removed after creation.
Like lists, tuples can store multiple values of different data types together, making them useful for grouping related data that should remain constant.

Tuples are often used in data science to represent fixed datasets, coordinates, or records where immutability is important for data integrity.
