## 1. If–Elif–Else (Conditional Statements)

** Basic Structure:

if: Checks the main condition

elif: Checks additional conditions

else: Executes if no condition is met

```python
score = 85

if score >= 90:
    print("Excellent performance.")
elif score >= 70:
    print("Good performance.")
else:
    print("Needs improvement.")
```

```python
x = int(input("Enter a number: "))

if x % 2 == 0:
    print("Your number is even")

print("Your program has ended")
```

```python
x = int(input("Enter a number: "))

if x % 2 == 0:
    print("Your number is even")
else:
    print("Your number is odd")

print("Your program has ended")
```

```python
x = int(input("Enter a number: "))

if x < 10:
    print("The number is less than 10")
    
elif x == 10:
    print("The number is equal to 10")
    
else:
    print("The number is greater than 10")
    
print("Your program has ended")
```

```python
x = int(input("Enter a number between 0 and 100: "))

if x == 100:
    print("The number is 100")

elif x >= 90:
    print("The number is between 90 and 100")

elif x >= 80:
    print("The number is between 80 and 90")
    
else:
    print("The number is less than 80")
    
print("Your program has ended")
```

```python
x = int(input("Enter a number: "))

if x % 3 == 0:
    if x % 2 == 0:
        print("The number is divisible by both 2 and 3")
    else:
        print("The number is divisible by 3 but not by 2")
else:
    print("The number is not divisible by 3")
    
print("Your program has ended")
```
